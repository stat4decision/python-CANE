Représentez plusierus nuages de points sur le même graphique et modifiez les axes


```python
plt.plot(np.random.random(100),np.random.random(100),".",c="red",label="A")
plt.plot(np.random.random(100),np.random.random(100),".",c="blue",label="B")
plt.title("Mon graphique")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(bbox_to_anchor=(1.2,0.5))
plt.savefig("mon_graphique.jpg",bbox_inches="tight",pad_inches=0.5)
#plt.show()
```

A partir d’un jeu de données, représentez un nuage de points basé sur deux variables continues


```python
frame_housing=pd.read_csv("../data/housing.csv")
```

```python
frame_housing.head()
```

```python
frame_housing.shape
```

```python
#plt.scatter(frame_housing["housing_median_age"],frame_housing["median_income"])

plt.scatter("median_house_value","median_income",data=frame_housing, s=1,
            c="housing_median_age")
plt.xlabel("median_house_value")
plt.ylabel("median_income")
plt.colorbar()
```

Générez un histogramme à partir des données


```python
plt.hist("population",data=frame_housing,bins=50,label="pop",alpha=0.5)
plt.hist("households",data=frame_housing,bins=50,label="house",alpha=0.5)
plt.legend()
```

```python
for labels in frame_housing["ocean_proximity"].unique():
    plt.hist(frame_housing[frame_housing["ocean_proximity"]==labels]["housing_median_age"],
             bins=50,label=labels,alpha=0.5)
plt.legend()
```

## Représentation sur une carte

plt.imshow et le paramètre extent=[-124.55, -113.80, 32.45, 42.05]
On sauve cette image


```python
plt.scatter("longitude","latitude",data=frame_housing, s=frame_housing["population"]/500,
            c="median_house_value",cmap=plt.get_cmap("rainbow"),alpha=0.8)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend(["Population"])
plt.colorbar().ax.set_ylabel('median house value') 

plt.imshow(plt.imread("../data/california.png"),extent=[-124.55, -113.80, 32.45, 42.05])
```
