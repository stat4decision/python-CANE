Générer un array de nombres aléatoires puis remplacer les valeurs plus petites que 0 par 0. Trier l’array obtenu


```python
array_rand=np.random.randn(100000)
array_rand.mean()
```

```python
array_rand=np.sort(np.where(array_rand<0,0,array_rand))
array_rand.mean()
```

```python
%%timeit
for i in range(len(array_rand)):
    if array_rand[i]<0:
        array_rand[i]=0
```

Calculer des statistiques sur les colonnes d'une matrice

```python
array2=np.random.random((1000,10))
```
